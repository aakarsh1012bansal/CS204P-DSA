```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
df
```

```python
df.TypeName.value_counts()
```

```python
df.Product.value_counts()
```

```python
df.drop('Product', axis=1, inplace=True)
df
```

```python
df.Company.value_counts()
```

```python
df.OpSys.value_counts()
```

```python
pd.get_dummies(df['Company'], dtype=int)
pd.get_dummies(df['TypeName'], dtype=int)
pd.get_dummies(df['OpSys'], dtype=int)
```

```python
df = df.join(pd.get_dummies(df['Company'], dtype=int))
df = df.join(pd.get_dummies(df['TypeName'], dtype=int))
df = df.join(pd.get_dummies(df['OpSys'], dtype=int))
df.drop(['Company', 'TypeName', 'OpSys'], axis=1, inplace=True)
df
```

```python
df.ScreenResolution.str.split(' ')
res = df.ScreenResolution.str.split(' ').apply(lambda x: x[-1])
res.str.split('x')
ScreenX = res.str.split('x').apply(lambda x: x[0])
ScreenY = res.str.split('x').apply(lambda x: x[1])
ScreenX = ScreenX.astype('int')
ScreenY = ScreenY.astype('int')
df.drop('ScreenResolution', axis=1, inplace=True)
df.insert(2, 'ScreenX', ScreenX)
df.insert(3, 'ScreenY', ScreenY)
df
```

```python
df.Cpu.str.split(' ')
df.Cpu.value_counts()
cpub = df.Cpu.str.split(' ').apply(lambda x: x[0])
cpuf = df.Cpu.str.split(' ').apply(lambda x: x[-1].replace('GHz', ''))
cpuf = cpuf.astype('float')
df.drop('Cpu', axis=1, inplace=True)
df.insert(2, 'CPU_Brand', cpub)
df.insert(3, 'CPU_Frequency', cpuf)
df
```

```python
df.CPU_Brand.value_counts()
```

```python
df.Ram.value_counts()
df['Ram'] = df.Ram.str.replace('GB', '').astype('int')
df
```

```python
def TBtoGB(value):
    if value[-2:] == 'TB':
        return float(value[:-2]) * 1024
    else:
        return float(value[:-2])
```

```python
mem = df.Memory.str.split(' ').apply(lambda x: x[0])
memtype = df.Memory.str.split(' ').apply(lambda x: x[1])
# memtype.value_counts()
mem = mem.apply(lambda x: TBtoGB(x))
mem.value_counts()
```

```python
df.drop('Memory', axis=1, inplace=True)
df.insert(2, 'Memory', mem.astype('int'))
df.insert(3, 'Memory_Type', memtype.astype('str'))
df
```

```python
df.columns
```

```python
df['GPU'] = df.Gpu.str.split(' ').apply(lambda x: x[0])
df.drop('Gpu', axis=1, inplace=True)
df
```

```python
df['Weight'] = df.Weight.str.replace('kg', '').astype('float')
df
```

```python
df = df.join(pd.get_dummies(df['Memory_Type'], dtype=int))
df
```

```python
df.drop('Memory_Type', axis=1, inplace=True)
```

```python
df
```

```python
df.columns
```

```python
gpuCategories = pd.get_dummies(df['GPU'], dtype=int)
cpuCategories  = pd.get_dummies(df['CPU_Brand'], dtype=int)
gpuCategories.columns= [str(x) + '_GPU' for x in gpuCategories.columns]
cpuCategories.columns= [str(x) + '_CPU' for x in cpuCategories.columns]
df = df.join(gpuCategories)
df = df.join(cpuCategories)
df.drop(['GPU', 'CPU_Brand'], axis=1, inplace=True)
df
```

```python
df.to_csv('laptop_price_cleaned.csv', index=False)
```

```python
df.corr()
```

```python
allfeatures = df.corr()['Price_euros'].apply(lambda x: abs(x)).sort_values(ascending=False)
allfeatures
```

```python
selectedfeatures = allfeatures[:26].index
```

```python
finalDF = df[selectedfeatures]
finalDF
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
x, y = finalDF.drop('Price_euros', axis=1), finalDF.Price_euros
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
